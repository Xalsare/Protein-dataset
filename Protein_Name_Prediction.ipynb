```python
accession_numbers = ["Q99928"]


for accession_number in accession_numbers:
    url = f"https://rest.uniprot.org/uniprotkb/{accession_number}.txt"
    response = requests.get(url)
    if response.status_code == 200:
      print(f"Data for {accession_number}:\n{response.text}\n")
    else:
        print(f"Failed to fetch data for {accession_number}. Status code: {response.status_code}")
```

```python
# Assuming 'protein_data' is your dataset containing protein accession numbers
accession_numbers = protein_data['Accession'].tolist()
protein_details = {'Accession': [], 'Data': []}

for accession_number in accession_numbers:
    url = f"https://www.uniprot.org/uniprot/{accession_number}.txt"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.text
        protein_details['Accession'].append(accession_number)
        protein_details['Data'].append(data)

        with open(f"{accession_number}.txt", "w") as file:
            file.write(data)
        print(f"Data for {accession_number} written to file.\n")
    else:
        print(f"Failed to fetch data for {accession_number}. Status code: {response.status_code}")

    time.sleep(1)

# Ensure all lists have the same length
max_length = max(len(protein_details['Accession']), len(protein_details['Data']))
protein_details['Accession'] += [''] * (max_length - len(protein_details['Accession']))
protein_details['Data'] += [''] * (max_length - len(protein_details['Data']))

# Convert protein_details to DataFrame
protein_details_df = pd.DataFrame(protein_details)

# Merge the original dataset with the protein_details_df on 'Accession' column
merged_data = pd.merge(protein_data, protein_details_df, on='Accession', how='outer')

# Save the updated dataset
merged_data.to_csv('updated_protein_data.csv', index=False)
```
